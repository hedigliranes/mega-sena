```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
# Atribuir o nome da planilha para a variável "arquivo"
arquivo = 'basedados.xlsx'
# Carregar a base de dados para um objeto pandas.io.excel.ExcelFile
excel = pd.ExcelFile(arquivo)
# Carregar a planilha excel para um objeto dataframe
dados = excel.parse(0)
```

Sem dúvidas um dos grandes sonhos de qualquer brasileiro é ganhar na mega-sena, milhões e milhões de reais sorteados periódicamente, porém qual a real chance de ganhar ? Quando pegamos a quadra como exemplo a chance de alguém acertar 4 de 60 números é de 1/2.332, ou seja , já se pode considerar complicado acerta apenas 4 números, mas se subirmos para 5 ? Para ganhar na quina a chance de alguém ter a sequência de 5 números sorteada é de 1/154.518, olhando parece algo extremamente complicado, mas e a mega-sena ? Para se acertar 6 números em 60 a chance é 1/50.063.860, práticamente impossível,mas só teóricamente, pois sempre existe um sortudo ou dois que eventualmente acertam a tão sonhada sequência, mas seria sorte ? estratégia ? vamos olhar pro passado, talvez tentando os números que mais sairam na história da mega-sena. 

```python
list_values = []
most_numberA = dados.iloc[:,2:8].unstack().value_counts()[0:6].index
most_numbers = dados.iloc[:,2:8].unstack().value_counts().to_dict()
values = most_numbers.values()
for i in values:
    list_values.append(i)
list_values = list_values[0:6]
```

```python
fig1, ax1 = plt.subplots()
labels = most_numberA[0],most_numberA[1],most_numberA[2],most_numberA[3],most_numberA[4],most_numberA[5]
explode = (0.1, 0, 0, 0,0,0)
ax1.pie(list_values, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

Podemos ver através do gráfico os seis números que mais foram sorteados ao longo dos anos na mega-sena, onde
vemos que o número 5 é o mais sorteado e sua diferença para o número 23, o sexto mais sorteado, é de menos de 1%,
ou seja, a incidência desses números em sorteios é bastante alta, assim para Senas, Quinas e Quadras eles podem 
acabar sendo fortes candidatos a sairem. Mas se olharmos os ganhadores ? 

```python
ganhadores = (dados.pivot_table(index=["Ganhadores_Sena"], values="Concurso", 
                   aggfunc="count")).index.values
ganhadores_values = (dados.pivot_table(index=["Ganhadores_Sena"], values="Concurso", 
                   aggfunc="count"))['Concurso'].to_dict()
```

```python
list_values2 = []
values2 = ganhadores_values.values()
for i in values2:
    list_values2.append(i)
```

```python
plt.plot(ganhadores,list_values2,color='green', marker='o', linestyle='dashed',
        linewidth=2, markersize=6)
plt.show()
```

Aqui vemos o quanto é complicado ganhar na mega sena, pois de todas as edições quase 1600 terminaram sem um vencedor e aproximadamente 400 com apenas um sortudo. Porém contra a maré temos alguns casos raros onde 15 e 17 pessoal dividiram o prêmio milionário. Sorte ? possível, mas quais as chances de 15 ou 17 pessoas jogarem exatamente os mesmos números e esse serem sorteados ? Vamos ver, se para que apenas você consiga ganhar as chances são de 1/50.063.860, se mais alguem ganhasse com você a chance disso ocorrer seria de (1/50.063.860)^2, a de 15 ganharem juntas (1/50.063.860)^15 e a de 17 seria (1/50.063.860)^17, ou seja , ou eles tem muita sorte ou esse sorteio sofreu uma fraude, mas isso nunca saberemos.
